```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("UqmxSipJO7BDQgI4") as mp:
    structure = mp.get_structure_by_material_id("mp-11520")
```

# DIFFRACTION D'UN RAYONNEMENT CuKA SUR LE CRISTAL YNiSb

```python
xrd = XRDCalculator('CuKa',1.54060)
pattern = xrd.get_pattern(structure, two_theta_range=(0,45))
xrd.show_plot(structure,annotate_peaks=False, two_theta_range=(0,45))
```

Premier pic:

```python

theta = pattern.x[0]
intensity = pattern.y[0]
hkl = pattern.hkls[0]

print("Le premier pic se situe à 2θ = " + str(theta)+"° avec une intensité de " + str(intensity)+" et les indices hkl associés sont " +str(hkl[0]['hkl'])+"et de multiplicité " + str(hkl[0]['multiplicity'])) 
```
